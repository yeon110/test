```python
file='/content/drive/MyDrive/공공데이터인턴/국내1_end.xlsx'
```

```python
import pandas as pd
df=pd.read_excel(file)
df.head()
```

```python
cas=df['CAS번호'].tolist()
molecular=df['분자량'].tolist()
idlh=df['IDLH'].tolist()
pac=df['PAC-2'].tolist()
```

```python
len(df)
```

```python
mole=pd.DataFrame({'cas':cas,'molecular weight':molecular,'IDLH':idlh,'pac-2':pac})
```

```python
molecular_c=molecular.copy()
idlh_c=idlh.copy()
pac_c=pac.copy()
```

```python
mole.head()
```

```python
#idlh, pac-2-국내
sum=0
idlh_f=[]
pac_f=[]

for i in range(len(idlh_c)):
  if idlh_c[i] == '-' or pac_c[i] == '-':
    idlh_f.append(idlh_c[i])
    pac_f.append(pac_c[i])
  elif mole.iloc[i,2] == 'HSDB':
    idlh_f.append('HSDB')
    pac_f.append(mole.iloc[i,3].split('·')[1])
  else:
    idlh_f.append(mole.iloc[i,2].split('·')[1])
    pac_f.append(mole.iloc[i,3].split('·')[1])

```

```python
print(len(idlh_f))
print(len(pac_f))
```

```python
#분자량-국내
import re
mole_f=[]
for i in range(len(molecular_c)):
  m=molecular_c[i]
  if m == '-':
    mole_f.append(m)
  else:
    mole_f1=m.split('·')[1]
    
    if '-' or '~' in m:
      mole_f2=re.split('[-~]',mole_f1)
      mole_f.append(mole_f2[0])
    else:
      mole_f.append(mole_f1)
  

print(len(mole_f))

```

```python
mole_df=pd.DataFrame({'CAS번호':cas,'molecular weight':mole_f,'IDLH':idlh_f,'PAC-2':pac_f})
```

```python
mole_df
```

```python
mole_df.iloc[0,3]
```

```python
pac_cal=[]
idlh_cal=[]
for i in range(len(mole_df)):

  if 'ppm' in mole_df.iloc[i,3]:
    p=mole_df.iloc[i,3]
    p=p.strip()
    if '*' in p:
      p=p.split('*')[0]
    else:
      p=p.split('ppm')[0]
    pac_cal.append(str(p)+' ppm')

  elif 'mg' in mole_df.iloc[i,3]:
    a=mole_df.iloc[i,3]
    a=a.strip()
    mg=a.split('mg')[0]
    
    w=mole_df.iloc[i,1]
    wei=w.strip()
    
    try:
      if float(wei) == 0:
        pac_cal.append('-')
      else:
        ppm=int(float(mg)*24.45/float(wei))
        z=str(ppm)+' ppm'
        z=z.strip()
        pac_cal.append(z)
      
    except ValueError:
      c=mole_df.iloc[i,3]
      c=c.strip()
      c=str(c.split('mg')[0])+'mg/m3'
      pac_cal.append(c)
  
  else:
    pac_cal.append('-')
  
  if 'ppm' in mole_df.iloc[i,2]:
    m=mole_df.iloc[i,2]
    m=m.strip()
    if '*' in m:
      m=m.split('*')[0]
    else:
      m=m.split('ppm')[0]
    idlh_cal.append(str(m)+' ppm')

  elif 'mg' in mole_df.iloc[i,2]:
    b=mole_df.iloc[i,2]
    b=b.strip()
    mg=b.split('mg')[0]
    
    try:
      if float(wei) == 0:
        idlh_cal.append('-')
      else:
        ppm=int(float(mg)*24.45/float(wei))
        z=str(ppm)+' ppm'
        z=z.strip()
        idlh_cal.append(z)
      
    except ValueError:
      c=mole_df.iloc[i,2]
      c=str(c.split('mg')[0])+'mg/m3'
      c=c.strip()
      idlh_cal.append(c)
  else:
    idlh_cal.append('-')
```

```python
print(len(pac_cal))
print(len(idlh_cal))
```
